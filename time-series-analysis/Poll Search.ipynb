```python
start_date = datetime.datetime(2017, 5, 7, 0, 0)
end_date = datetime.datetime(2019, 4, 30, 0, 0)
approval_rating_df = approval_rating_df.loc[(approval_rating_df['enddate'] >= start_date) & (approval_rating_df['enddate'] <= end_date)]
```

```python
def find_longest_contiguous(df):
    max_length = 0
    for index, row in df.iterrows():
        curr = 0
        start_date = row['enddate']
        curr_date = start_date
        new_df = df.loc[df['enddate'] > start_date]
        for index, nested_row in new_df.iterrows():
            diff = (nested_row['enddate']-curr_date).days
            if diff != 1:
                break
                
            curr += 1
            if curr > max_length:
                max_length = curr
                max_start_date = start_date
                max_end_date = curr_date
            curr_date += datetime.timedelta(days=1)
    print("found max length of ", max_length, " starting at ", max_start_date, " ending at ", max_end_date)
```

```python
find_longest_contiguous(you_gov_df)
```

```python
gallup_df = approval_rating_df.loc[(approval_rating_df['pollster'] == 'Gallup') & (approval_rating_df['subgroup'] == 'All polls')]
find_longest_contiguous(gallup_df)
```

```python
morning_consult_df = approval_rating_df.loc[(approval_rating_df['pollster'] == 'Morning Consult')]
find_longest_contiguous(morning_consult_df)
```
